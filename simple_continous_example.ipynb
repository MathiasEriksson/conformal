# Simple conformal prediction example in the *continous* regression setting

```python
import numpy as np
import torch
import pyro
import pyro.contrib.gp as gp
import pyro.distributions as dist
import matplotlib.pyplot as plt

pyro.set_rng_seed(0)
torch.set_default_dtype(torch.double)
```

### Data generating process - homoskedastic noise

```python
pyro.set_rng_seed(0)
N = 100
noise_level = 0.05
x = torch.linspace(-5.0,5.0,N)
support = torch.linspace(-5.0,5.0,1000)
a = torch.tensor(3)
y = torch.sinc( a * support )
y_obs = torch.sinc( a * x ) + dist.Normal(0.0, noise_level).sample(sample_shape=(N,))
plt.plot(x, y_obs, '+')
plt.plot(support, y)
plt.show()
```
